#W6A9

```python
def read_dict():
    d={}
    s=input().strip()
    if s=="":
        return{}
    s=s[1:-1]
    if s=="":
        return{}
    items=s.strip().split(',')
    for item in items:
        key,value=item.split(":")
        key=key.strip().strip("'")
        value=int(value.strip())
        d[key]=value
    return d

d=read_dict()
e=read_dict()
f={}

for x in d:
    if x in e:
        f[x]=d[x] +e[x]
    else:
        f[x]=d[x]
for x in e:
    if x not in d:
        f[x]=e[x]

last='{'
l1st=[]
for z in sorted(f):
    l1st.append(f"'{z}': {f[z]}")
last+= ",".join(l1st) + '}'
print(last)
```

#W6A10

```python
def sum(a,target):
    dic={}
    result=[]
    for x,y in enumerate(a):
        find=target-y
        if find in dic:
            result.append((find, y))
        dic[y]=x

    return sorted(result)

a=list(map(int,input().split()))
b=int(input())
print(sum(a,b))
```
