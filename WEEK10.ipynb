#W10A3

```python
def sohp(b):
    s = 0
    for x in str(b):
        s += int(x) ** 2
    return s

def sohpriel(b):
    seen = set()
    c = b
    while c != 1:
        if c in seen:
            return False
        seen.add(c)
        c = sohp(c)
    return True

try:
    a = int(input())
    if sohpriel(a):
        print("YES")
    else:
        print("NO")
except ValueError:
    print("Loi gia tri")
```

```python
try:
    n=int(input())
    res=[]
    d={}
    while n>0:
        x=int(input())
        res.append(x)
        n-=1
    for x in res:
        if x not in d:
            d[x]=1
        else:
            d[x]+=1
    max=0
    for x in d:
        if d[x]>max:
            max=d[x]
    if max==1:
        sum=0
        for i in res:
            sum+=i
        print(sum)
    else:
        raise ValueError
except ValueError:
    print("Mang khong hop le")
```

```python
import re

t = int(input())
for _ in range(t):
    s = input().strip()
    try:
        re.compile(s)
        print("True")
    except re.error:
        print("False")
```

```python
def is_valid_regex(s):
    operators = {'*', '+', '?'}
    n = len(s)
    i = 0

    while i < n:
        if s[i] == '\\':
            if i == n - 1:
                return False
            i += 2
            continue

        if s[i] in operators:
            if i == 0:
                return False
            if s[i - 1] in operators:
                return False
        i += 1
    return True


t = int(input())
for _ in range(t):
    s = input().strip()
    print("True" if is_valid_regex(s) else "False")
```
