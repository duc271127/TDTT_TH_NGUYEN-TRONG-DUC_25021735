```python
#Sbm=sxq + 2 s day
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        self.pi = 3.14

    def surface_area(self):
        return 2 * self.pi * self.radius * (self.radius + self.height)

    def volume(self):
        return self.pi * self.radius ** 2 * self.height

r, h = map(float, input().split())

cylinder = Cylinder(r, h)

area = cylinder.surface_area()
vol = cylinder.volume()

print(f"{area:.2f} {vol:.2f}")
```

```python
class Date:
    def __init__(self, d, m, y):
        self.d = d
        self.m = m
        self.y = y

    def is_leap_year(self):
        return self.y % 400 == 0 or (self.y % 4 == 0 and self.y % 100 != 0)

    def is_valid(self):
        if self.y <= 0 or self.m < 1 or self.m > 12 or self.d < 1:
            return False

        date = [31, 28, 31, 30, 31, 30,
                         31, 31, 30, 31, 30, 31]

        if self.m == 2 and self.is_leap_year():
            return self.d <= 29

        return self.d <= date[self.m - 1]

    def next_day(self):
        date = [31, 28, 31, 30, 31, 30,
                         31, 31, 30, 31, 30, 31]

        if self.m == 2 and self.is_leap_year():
            max_day = 29
        else:
            max_day = date[self.m - 1]

        if self.d < max_day:
            self.d += 1
        else:
            self.d = 1
            if self.m < 12:
                self.m += 1
            else:
                self.m = 1
                self.y += 1

    def __str__(self):
        return f"{self.d:02d}/{self.m:02d}/{self.y}"


s = input().strip()

try:
    d, m, y = map(int, s.split('/'))
    date = Date(d, m, y)

    if not date.is_valid():
        print("INVALID")
    else:
        date.next_day()
        print(date)

except:
    print("INVALID")
```

```python
class ShoppingCart:
    def __init__(self):
        self.gio=[]

    def them_hang(self,name,price):
        self.gio.append({"name":name, "price":price})
    
    def xoa_hang(self,name):
        for sp in self.gio:
            if sp["name"]==name:
                self.gio.remove(sp)
            return True
        return False
    
    def gio_rong(self):
        return len(self.gio)==0
    
    def tong_tien(self):
        tong=0
        for sp in self.gio:
            tong+=sp["price"]
        return tong
    
    def xem_sp(self):
        if self.gio_rong():
            print("Giỏ hàng trống")
        else:
            for i, product in enumerate(self.gio, 1):
                print(f"{i}. {product['name']} - {product['price']}")

    def xoa_sach(self):
        self.gio.clear()

cart = ShoppingCart()

while True:
    print("\n----- MENU -----")
    print("2. Thêm sản phẩm")
    print("3. Xóa sản phẩm theo tên")
    print("4. Kiểm tra giỏ hàng rỗng")
    print("5. Tính tổng tiền")
    print("6. Hiển thị giỏ hàng")
    print("7. Xóa toàn bộ giỏ hàng")

    choice = input("Nhập lựa chọn (2-7): ")

    if choice == '2':
        name = input("Nhập tên sản phẩm: ")
        price = float(input("Nhập giá sản phẩm: "))
        cart.them_hang(name, price)

    elif choice == '3':
        name = input("Nhập tên sản phẩm cần xóa: ")
        if cart.xoa_hang(name):
            print("Đã xóa sản phẩm")
        else:
            print("Không tìm thấy sản phẩm")

    elif choice == '4':
        if cart.gio_rong():
            print("Giỏ hàng đang rỗng")
        else:
            print("Giỏ hàng không rỗng")

    elif choice == '5':
        print("Tổng tiền:", cart.tong_tien())

    elif choice == '6':
        cart.xem_sp()

    elif choice == '7':
        cart.xoa_sach()
        print("Đã xóa toàn bộ giỏ hàng")

    else:
        print("Lựa chọn không hợp lệ")

    stop = input("Có muốn thoát khỏi chương trình không? ")
    if stop.lower() == "yes":
        break



            


    
```

```python
import math
class Fraction:
    def __init__(self,num,den):
        if den==0:
            raise ValueError("Mẫu số phải khác 0")
        if den <0:
            den=-den
            num=-num
        uoc=math.gcd(num,den)
        self.num=num//uoc
        self.den=den//uoc
    
    def add(self,other):
        return Fraction(
            self.num*other.den + other.num* self.den,
            self.den*other.den
        )
    
    def sub(self,other):
        return Fraction(
            self.num*other.den - other.num* self.den,
            self.den*other.den
        )
    def mul(self,other):
        return Fraction(
            self.num  * other.num,
            self.den * other.den
        )
    
    def div(self,other):
        return Fraction(
            self.num * other.den,
            self.den * other.num
        )
    
    def __str__(self):
        return f'{self.num}/{self.den}'
    
a,b,op,c,d=input().split()
a,b,c,d=int(a),int(b),int(c),int(d)
a1=Fraction(a,b)
a2=Fraction(c,d)

if op=="+":
    print(f"{a1.add(a2)}")
if op=="-":
    print(f'{a1.sub(a2)}')
if op=="*":
    print(f'{a1.mul(a2)}')
if op=="/":
    print(f'{a1.div(a2)}')
    
```

```python
class BankAccount:
    def __init__(self,owner,balance):
        self.owner=owner
        self.balance=balance
    
    def deposit(self,num):
        self.balance += num 
    def withdraw(self,num):
        if int(self.balance) - int(num) >=0:
            self.balance-=num
    def __str__(self):
        return f"{self.owner} : {self.balance}"

a,b=input().split()
b=int(b)
c=BankAccount(a,b)
n=int(input())
while n>0:
    lenh,gia=input().split()
    if lenh=="DEPOSIT":
        c.deposit(int(gia))
    elif lenh=="WITHDRAW":
        c.withdraw(int(gia))
    n-=1
print(f"{c.balance:.2f}")
```

```python
class Book:
    def __init__(self,title,author,year):
        self.title=title
        self.author=author
        self.year=year

class Library:
    def __init__(self):
        self.books=[]
    
    def add_book(self, book):
        self.books.append(book)
    
    def count_by_author(self,name):
        count=0
        for book in self.books:
            if book.author==name:
                count+=1
        return count
    
    def find_by_year(self,year):
        count=0
        for book in self.books:
            if book.year==year:
                count+=1 
        return count

a=int(input())
c=Library()

for i in range(a):
    b=input().strip()

    if b[:3]=="ADD":
        data=b[4:]
        title,author,year=data.split(";")
        year=int(year)
        book=Book(title,author,year)
        c.add_book(book)
    elif b[:9]=="COUNTYEAR":
        year=int(b.split()[1])
        print(c.find_by_year(year))
    elif b[:5]=="COUNT":
        author=b[6:]
        print(c.count_by_author(author))
```

```python
import math

class Point:
    def __init__(self,x,y,z):
        self.x=x
        self.y=y
        self.z=z
    
    def vector(self,other): #quy ve vector
        return(
            other.x -self.x,
            other.y -self.y,
            other.z -self.z
        )
    def cross(u,v): #phap tuyen
        return(
            u[1]*v[2] - u[2] *v[1],
            u[2]*v[0] -u[0]*v[2],
            u[0]*v[1] -u[1]*v[0]
        )
    def dot(u,v): # tich vo huong
        return u[0]*v[0] +u[1] *v[1] +u[2]*v[2]
    def norm(u): # do dai
        return math.sqrt(Point.dot(u,u))
    
A = Point(*map(float, input().split()))
B = Point(*map(float, input().split()))
C = Point(*map(float, input().split()))
D = Point(*map(float, input().split()))

AB = A.vector(B)
AC = A.vector(C)
BC = B.vector(C)
BD = B.vector(D)

n1 = Point.cross(AB, AC)
n2 = Point.cross(BC, BD)

cos_phi = abs(Point.dot(n1, n2)) / (Point.norm(n1) * Point.norm(n2))
phi = math.degrees(math.acos(cos_phi))

print(f"{phi:.2f}")
```

```python
import math
class Triangle:
    def __init__(self,initx,inity,initz):
        self.x=initx
        self.y=inity
        self.z=initz
    def get_Area(self):
        if self.x+self.y>self.z and self.x + self.z> self.y and self.y+self.z > self.x:
            p=(self.x + self.y +self.z)/2
            area=math.sqrt(p*(p-self.x)*(p-self.y) * (p-self.z))
            return area
        return -1
    
a,b,c=map(int ,input().split())
triangle=Triangle(a,b,c)
area=triangle.get_Area()

if area==-1:
    print("INVALID")
else:
    print(f'{area:.2f}')
```
