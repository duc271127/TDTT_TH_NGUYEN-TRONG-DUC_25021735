```python
def binary_search(a,target):
    start=0
    end=len(a)-1
    while start<=end:
        mid=(start+end)//2
        if a[mid] > target:
            end-=1
        elif a[mid] < target:
            start+=1
        else:
            return mid
    return -1

a=list(map(int,input().split()))
target=int(input())
print(binary_search(a,target))

```

```python
def bubble_sort(a):
    for i in range(1,len(a)):
        for j in range(0,len(a)-1):
            if a[j]>a[j+1]:
                a[j],a[j+1]=a[j+1],a[j]
    return a

a=list(map(int,input().split()))
print(bubble_sort(a))

```

```python
def selection_sort(a):
    for i in range(len(a)):
        min=i
        for j in range(i+1,len(a)):
            if a[j] < a[min]:
                min =j
        a[i],a[min]=a[min],a[i]
    return a

a=list(map(int,input().split()))
print(selection_sort(a))
```

#W7A8

```python
import heapq

def w7a8(intervals,queries):
    intervals.sort()
    
    minHeap=[]
    res,i={}, 0

    for q in sorted(queries):
        while i<len(intervals) and intervals[i][0] <= q:
            l,r= intervals[i]
            heapq.heappush(minHeap,(r-l+1, r))
            i+=1
        while len(minHeap)>0 and minHeap[0][1] < q: #while minHeap <=> while len(minHeap) >0
            heapq.heappop(minHeap)
        res[q] = minHeap[0][0] if minHeap else -1
    return [res[q] for q in queries]

a=eval(input())
b=eval(input())
print(w7a8(a,b))
```

```python
#CACH 2:BINARY SEARCH TREE

def binary_search(l1st,target):
    start=0
    end=len(l1st) -1
    while start<=end:
        mid=(start+end) //2
        if l1st[mid] < target:
            start+=1
        elif l1st[mid] > target:
            end-=1
        else:
            return mid
    return start
        
def w7a9(l1st):
    res=[]
    sorted=[]

    for x in l1st[::-1]:
        posi= binary_search(sorted,x)
        res.append(posi)

        sorted.insert(posi,x)
    return res[::-1]
a=eval(input())
print(w7a9(a))
```

```python
def w7a10(l1st,m):
    res=[]
    for i in range(len(l1st)):
        sum=0
        for j in range(i, len(l1st)):
            sum+=l1st[j]
            res.append(sum%m)
    return max(res)

a=eval(input())
b=int(input())
print(w7a10(a,b))
```

```python
def w7a10(nums, m):
    prefix_sum = 0          
    answer = 0              
    prefix_sorted = []      
    for num in nums:
        prefix_sum = (prefix_sum + num) % m
        answer = max(answer, prefix_sum)

        left = 0
        right = len(prefix_sorted)

        while left < right:
            mid = (left + right) // 2
            if prefix_sorted[mid] <= prefix_sum:
                left = mid + 1
            else:
                right = mid


        if left < len(prefix_sorted):
            candidate = (prefix_sum - prefix_sorted[left] + m) % m
            answer = max(answer, candidate)

        prefix_sorted.insert(left, prefix_sum)

    return answer
```

```python
def find_nemo(l1st):
    count=len(l1st)
    for i in range(len(l1st)):
        if l1st[i]=="Nemo":
            if i==0:
                return f"{l1st[len(l1st)]} and {l1st[i+1]}"
            elif i==count-1:
                return f"{l1st[i-1]} and {l1st[0]}"
            else:
                return f"{l1st[i-1]} and {l1st[i+1]}"
    return 'Not Found'

a=int(input())
b=[]
while a>0:
    c=input()
    b.append(c)
    a-=1
print(find_nemo(b))
```
