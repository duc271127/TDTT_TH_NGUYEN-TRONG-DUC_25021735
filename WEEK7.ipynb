```python
def binary_search(a,target):
    start=0
    end=len(a)-1
    while start<=end:
        mid=(start+end)//2
        if a[mid] > target:
            end-=1
        elif a[mid] < target:
            start+=1
        else:
            return mid
    return -1

a=list(map(int,input().split()))
target=int(input())
print(binary_search(a,target))

```

```python
def bubble_sort(a):
    for i in range(1,len(a)):
        for j in range(0,len(a)-1):
            if a[j]>a[j+1]:
                a[j],a[j+1]=a[j+1],a[j]
    return a

a=list(map(int,input().split()))
print(bubble_sort(a))

```

```python
def selection_sort(a):
    for i in range(len(a)):
        min=i
        for j in range(i+1,len(a)):
            if a[j] < a[min]:
                min =j
        a[i],a[min]=a[min],a[i]
    return a

a=list(map(int,input().split()))
print(selection_sort(a))
```

#W7A5

```python
def count_pair(a,target):
    count=0
    d={}
    for x in a:
        b=target-x
        if b in d:
            count+=d[b]
        d[x]=d.get(x,0) +1
    return count

a=list(map(int,input().split()))
target=int(input())
print(count_pair(a,target))
```

```python
def flatten(a):
    result=[]
    for x in a:
        for y in x:
            result.append(y)
    return result

a=eval(input())
print(flatten(a))
```

```python
def find(a):
    b=[]
    for i in range(len(a)):
        count=1
        for j in range(i,len(a)-1):
            if a[j] < a[j+1]:
                count+=1
        b.append(count)
    return max(b)

a=eval(input())
print(find(a))
```
