#W5A3

```python
def nt(n):
    if n<2:
        return False
    if n==2:
        return True
    if n>2:
        for i in range(2,int(n**1/2)):
            if n%i==0:
                return False
        return True

a=int (input())
print(nt(a))
```

#W5A4

```python
def shh(n):
    sum=0
    for i in range(1,n):
        if n%i==0:
            sum+=i
    if sum == n:
        return True
    else:
        return False
    
a=int(input())
print(shh(a))
```

#W5A5

```python
def checkychecky(a,k):
    for i in range(len(a)):
        if a[i]==k:
            return i+1
    return -1

a=list(map(int,input().split()))
k=int(input())
print(checkychecky(a,k))
```

#W5A8

```python
def hamming(a,b):
    list1=[]
    list2=[]

    while a!=0:
        list1.append(a%2)
        a//=2
    while b!=0:
        list2.append(b%2)
        b//=2
    
    list1=list1[::-1]
    list2=list2[::-1]

    if len(list1)>len(list2):
        list2=[0]*(len(list1)-len(list2)) + list2
    else:
        list1=[0] *(len(list2)-len(list1)) +list1
    
    count=0
    for i in range(len(list1)):
        if list1[i] != list2[i]:
            count+=1
    return count

a,b=map(int,input().split())
print(hamming(a,b))
```

#W5A10

```python
def dc(a,b):
    map_ab={}
    map_ba={}
    for x,y in zip(a,b):
        if x in map_ab and map_ab[x] !=y:
            return False
        if y in map_ba and map_ba[y] !=x:
            return False
        map_ab[x]=y
        map_ba[y]=x
    return True

d,c=map(str,input().split())
dc(d,c)
```
