```python
def areAnagrams(path1: str, path2: str) -> bool:
    with open(path1, 'r') as f:
        s1 = f.read().strip()
    with open(path2, 'r') as f:
        s2 = f.read().strip()
    
    return sorted(s1) == sorted(s2)


print(areAnagrams("input.txt","input2.txt"))
```

```python
def fibo_fake(n:int) -> str:
    if n==0:
        return 'a'
    elif n==1:
        return 'b'
    else:
        return fibo_fake(n-1) + fibo_fake(n-2)

def fibo(path:str) -> str:
    with open(path,'r') as f:
        n=int(f.read())
    return fibo_fake(n)
print(fibo("input.txt"))
```

#W11A12

```python
def maximumProduct(path:str) -> int:
    with open(path,'r') as f:
        n=f.readlines()
    res_fake=[]
    for x in range(1,len(n)):
        res_fake.append(int(n[x].strip()))
    res_riel=[]
    count=3
    while count>0:
        res_riel.append(max(res_fake))
        res_fake.remove(max(res_fake))
        count-=1
    product=1
    for x in res_riel:
        product*=x
    return product

print(maximumProduct("input.txt"))
    
```

#W11A13

```python
def moveZeroes(fileName:str)->list:
    with open(fileName,'r') as f:
        n=f.readlines()
    res=[]
    for x in n:
        res.append(int(x.strip()))
    for y in range(len(res)):
        if res[y]==0:
            res.pop(y)
            res.append(0)
    return res

print(moveZeroes("input.txt"))
```

#W11A14

```python
def averageTime(path:str) -> int:
    with open(path,'r') as f:
        n=f.readlines()
    time=[]
    for x in range(len(n)):
        y=n[x].split(",")
        time.append(float(y[1].strip()))
        if x==len(n)-1:
            count=y[0]
    sum_time=0
    for x in time:
        sum_time+=x
    average=sum_time/int(count)
    return round(average,2)
print(averageTime("input.txt"))
```

```python
def findMovies(path: str):
    with open(path,'r') as f:
        n = f.readlines()
    res = []
    for x in n:
        y = x.strip().split(',') 
        if float(y[3]) >= 8.0 and y[2] != "boring":
            res.append(y[1])
    if len(res) > 0:
        b=""
        for x in res:
            b+=x.strip() +"\n"
        return b

    else:
        return "All are boring"
print(findMovies("input.txt"))
```

#W11A16

```python
def chuvi(path:str) -> float:
    with open(path,'r') as f:
        n=f.readlines()
    area=0
    for x in range(0,len(n),2):
        y=n[x].strip().split(":")
        z=n[x+1].strip().split(":")
        x=z[1].split()
        if y[1].strip()=="SQUARE":
            area+=float(z[1])*2
        elif y[1].strip()=="RECTANGLE":
            area+= 2*(float(x[0]) + float(z[2]))
        elif y[1].strip()=="CIRCLE":
            area+=2*3.14*float(z[1])
    return area

print(chuvi("input.txt"))
```

#W11A17

```python
def productExceptSelf(path: str) -> list[int]:
    with open(path,'r') as f:
        nums = list(map(int, f.read().strip().split()))

    res_real = []

    for i in range(len(nums)):
        prod = 1
        for j in range(len(nums)):
            if i != j:
                prod *= nums[j]
        res_real.append(prod)
    return res_real
print(productExceptSelf("input.txt"))
```

#W11A18

```python
import math

def estimatedTime(path: str) -> float:
    with open(path, 'r') as f:
        lines = f.readlines()

    src = lines[0].split()
    x1 = float(src[2])
    y1 = float(src[4])

    des = lines[1].split()
    x2 = float(des[2])
    y2 = float(des[4])

    vel = lines[2].split()
    v = float(vel[1])

    distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

    time = distance / v

    return round(time, 2)

print(estimatedTime("input.txt"))
```

```python
def imageSmoother(path: str):
    with open(path, 'r') as f:
        img = [list(map(int, line.split())) for line in f]

    M = len(img)
    res = [[0]*M for _ in range(M)]

    for i in range(M):
        for j in range(M):
            s = 0
            cnt = 0
            for x in range(i-1, i+2):
                for y in range(j-1, j+2):
                    if 0 <= x < M and 0 <= y < M:
                        s += img[x][y]
                        cnt += 1
            res[i][j] = s // cnt

    return res

ans = imageSmoother("input.txt")
for row in ans:
    print(" ".join(map(str, row)))

```
